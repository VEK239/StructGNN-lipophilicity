This notebook counts the unique representations of molecules and shows the most common ones.
To run it, you need to save the custom representation of a molecule into a file.

```python
from ring_dictionary_creating import RingsDictionaryHolder
rings_dictionary_holder = RingsDictionaryHolder('rings_zinc_features.json')
data = rings_dictionary_holder.read_json_rings_dict()
```

```python
import pandas as pd
import seaborn as sns
```

```python
data_smiles = [*(data.keys())]
data_hash = [*(data.values())]
```

```python
values = pd.DataFrame({'smiles':data_smiles, 'hash':data_hash})
```
